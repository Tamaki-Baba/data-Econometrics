### 連続型確率変数の場合𝑝(𝑥)=1/5で、確率変数Xが0から5までの値をとるときの期待値と分散をpythonで計算

```python
# X が 0〜5 の値を取り、各値の確率が一様（1/6）とする
xs = list(range(6))
p = 1/6

# 期待値 E[X]
E = sum(x * p for x in xs)

# E[X^2]
E2 = sum((x**2) * p for x in xs)

# 分散 Var(X) = E[X^2] - (E[X])^2
Var = E2 - E**2

print("期待値 E[X] =", E)
print("分散 Var(X) =", Var)
```
