### 適切なベイジアンネットワークをプログラミングし、購入したときの年齢層、購入していないときの年齢層の確率をそれぞれもとめる。また、女性、40代、CM見た人がどれくらいの確率で商品を買うかを推定する。

```python
import pandas as pd

# データ読み込み
df = pd.read_csv("conversion_task.csv")

# 1. 購入・非購入ごとの年齢層確率

# 購入した人の年齢層分布
age_purchase = (
    df[df["purchase"] == 1]["age"]
    .value_counts(normalize=True)
    .sort_index()
)

# 購入していない人の年齢層分布
age_unpurchase = (
    df[df["purchase"] == 0]["age"]
    .value_counts(normalize=True)
    .sort_index()
)

print("購入時の年齢層確率")
print(age_purchase)
print("\n非購入時の年齢層確率")
print(age_unpurchase)


# 2. ベイジアン推定

# 女性・40代・CM視聴者が購入する確率

def conditional_prob(col, value, purchase_value):
    subset = df[df["purchase"] == purchase_value]
    return (subset[col] == value).mean()

# 事前確率
p_purchase = (df["purchase"] == 1).mean()
p_unpurchase = 1 - p_purchase

# 条件付き確率（尤度）
p_sex_female_given_purchase = conditional_prob("sex", 1, 1)
p_age_40_given_purchase = conditional_prob("age", 4, 1)
p_cm_seen_given_purchase = conditional_prob("CM", 1, 1)

p_sex_female_given_unpurchase = conditional_prob("sex", 1, 0)
p_age_40_given_unpurchase = conditional_prob("age", 4, 0)
p_cm_seen_given_unpurchase = conditional_prob("CM", 1, 0)

# ベイズの定理
numerator = (
    p_purchase
    * p_sex_female_given_purchase
    * p_age_40_given_purchase
    * p_cm_seen_given_purchase
)

denominator = numerator + (
    p_unpurchase
    * p_sex_female_given_unpurchase
    * p_age_40_given_unpurchase
    * p_cm_seen_given_unpurchase
)

posterior = numerator / denominator

print("\nP(購入 | 女性・40代・CM視聴) =", posterior)
```
